## ①ファイルから秘密鍵を読み込みます

```python
from Crypto.PublicKey import RSA
f = open('imported_private.pem','r')
key = RSA.import_key(f.read())
```

## ②秘密のメッセージを次のセルの「''」の間にコピー＆ペーストします

```python
msg = ''
```

## ③秘密鍵を使ってメッセージを復号します

```python
import codecs
from Crypto.Cipher import PKCS1_OAEP

def decrypt_rsa(prikey, ctext):
    decipher_rsa = PKCS1_OAEP.new(key)
    msg = decipher_rsa.decrypt(ctext).decode("utf-8")
    return msg

ciphertext = codecs.decode(msg, 'hex')
decrypt_rsa(key, ciphertext)
```
