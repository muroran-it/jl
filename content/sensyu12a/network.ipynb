# Network Utility

## DNS (DNS over HTTP)

```python
import json
from pyodide.http import open_url

def nslookup(domain, record_type="A"):
    url = f"https://dns.google/resolve?name={domain}&type={record_type}"
    data = json.loads(open_url(url).read())

    print(f"Name: {domain}")
    print(f"Type: {record_type}")

    for ans in data.get("Answer", []):
        print(ans["data"])

```

```python
nslookup("openai.com")
```

## ping (http responce time)

```python
import time
from pyodide.http import open_url

def http_ping(url):
    start = time.perf_counter()
    open_url(url).read()
    end = time.perf_counter()
    print(f"{url}: {(end - start) * 1000:.1f} ms")
```

```python
http_ping("https://example.com")
```
